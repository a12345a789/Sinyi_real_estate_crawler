```python
import time
#測試新的架構
start_time=time.time() #起始執行時間
#社區名稱
community=[]
#社區地址
address=[]
#屋齡
houseage=[]
#建物類型
housetype=[]
#社區id
community=[]
#社區list
com_list=[]

for page in tqdm(range(0,219)):
    time.sleep(10)
    url=f'https://www.sinyi.com.tw/communitylist/NewTaipei-city/hotdeal-desc/'+format(page)

    for i in url:
        init_page = requests.get(url,headers=headers).content
        init_soup = BeautifulSoup(init_page, 'lxml')
        ids = [tag['id'] for tag in init_soup.select('div[id]')]
        for idd in ids: #處理社區名稱id
            if (len(idd)==7) & (idd!='fakeBuy') :
                com_list.append(idd)
        aa = init_soup.find_all('div',class_="LongInfoCard_Type_Name") #社區名稱
        ad = init_soup.find_all('div',class_="LongInfoCard_Type_Address") #社區地址
        ae = init_soup.find_all('div',class_="LongInfoCard_Type_HouseInfo") #社區房屋相關資訊
        af = init_soup.find_all('div',class_="LongInfoCard_Type_Parking") # 總層數
        #處理社區名稱
        for j in aa:
            bb=j.getText().split('｜')[0]
            community.append(bb)
        #處理地址    
        for k in ad:
            add=k.getText().split('|')[0]
            address.append(add)
        #處理屋齡
#         for l in ae:
#             age=l.getText().split('|')[0]
#             age=re.sub('[\u4e00-\u9fa5]','',age)
#             houseage.append(age)
        #處理建物類型
        
        
end_time=time.time()
print("總執行時間為{total_time}分".format(total_time=(end_time-start_time)/60))
```

```python
dict_com={'name':community,
         'address':address,
         'age':houseage}
df=pd.DataFrame.from_dict(dict_com, orient='index') #若不加from_dict會出現array長度的錯誤
df=df.T
print(df.shape)
df.drop_duplicates(inplace=True)
df.head(100)
```

```python
#將分段地址切割出來
df['address1']=df['address'].str.split(',',expand=True)[0]
df['address2']=df['address'].str.split(',',expand=True)[1]
df['address3']=df['address'].str.split(',',expand=True)[2]
df['address4']=df['address'].str.split(',',expand=True)[3]
```

```python
df.drop(columns=['address'],inplace=True)
df.drop(columns=['age'],inplace=True)
df.drop_duplicates(inplace=True)
```
